# Substituindo pela Mediana

```python
imputerMedian = SimpleImputer(missing_values = np.NaN, strategy = 'median')
```

```python
imputerMedian.fit_transform(df[['R&D Spend']])
```

# Substituindo pela Moda ou Mais frequente

```python
imputerMedian = SimpleImputer(missing_values = np.NaN, strategy = 'most_frequent')
```

```python
imputerMedian.fit_transform(df[['R&D Spend']])
```

### Substituindo por um valor constante

```python
imputerMedian = SimpleImputer(missing_values = np.NaN, strategy = 'constant', fill_value = 14500)
```

```python
imputerMedian.fit_transform(df[['R&D Spend']])
```
